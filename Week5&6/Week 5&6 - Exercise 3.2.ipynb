```python
#load necessary libraries
import pandas as pd
import numpy as np
import squarify
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Load data into a dataframe
unemployement = pd.read_csv('unemployement-rate-1948-2010.csv')
```

```python
#check the dimension of the table/look at the data
unemployement.shape
```

```python
# Tree chart

# Set labels
a = unemployement.groupby('Year')[['Value']].sum().index.get_level_values(0).tolist()

# Set sizes
d = unemployement.groupby('Year')[['Value']].sum().reset_index().Value.values.tolist()

# Plot tree map
plt.figure(figsize=(15, 12))
squarify.plot(sizes=d, label=a, alpha=.7)
plt.axis('off')
plt.title("Python Tree Chart") 
plt.show()
```

```python
# Area Chart
plt.fill_between(unemployement.Year, unemployement.Value)
plt.title("Python Area Chart - Unemployment Rate by Year") 
plt.xlabel("Year") 
plt.ylabel("Value") 
#Load data into a dataframe
unemployement = pd.read_csv('unemployement-rate-1948-2010.csv')plt.show()
```

```python
#Load data into a dataframe
crashes = pd.read_excel('Air_Bus_Rail_Crashes.xlsx')
```

```python
#Stacked area chart
plt.stackplot(crashes.Year,crashes.Bus,crashes.Railroad,crashes.Air,)
plt.title("Python Stacked Chart - Bus, Railroad and Air Crashes by Year") 
plt.xlabel("Year")
plt.legend(["Bus","Railroad", "Air"],
           bbox_to_anchor = (1.25,1))
plt.show()
```

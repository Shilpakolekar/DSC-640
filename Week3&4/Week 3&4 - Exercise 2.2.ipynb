```python
#load necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Load data into a dataframe
world_population = pd.read_excel('world-population.xlsm')
```

```python
#check the dimension of the table/look at the data
world_population.shape
```

```python
# Line chart
plt.plot(world_population.Year, world_population.Population)
plt.title("Python Line Chart - World Population by Year") 
plt.xlabel("Year") 
plt.ylabel("Population") 
```

```python
# Step chart
plt.step(world_population.Year, world_population.Population)
plt.title("Python Step Chart - World Population by Year") 
plt.xlabel("Year") 
plt.ylabel("Population") 
```
